```python
#! pip install locales
import pandas as pd
import matplotlib.pyplot as plt
```

```python

dict_faturamento = {'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01'
    ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        
    ]}
```

```python
df = pd.DataFrame(dict_faturamento)

media_faturamento = df.valor.mean().round(2)
print(f'A média do faturamento é: {media_faturamento}')

ax = df.plot.bar(x='data_ref', y='valor', rot=0, figsize=(15,6))
ax.bar_label(ax.containers[0])
#plt.axis([df.valor.min, df.valor.max])

# def val_max(df):
#    lim_max = df.valor.max()*1.1
#    return lim_max

plt.title('Faturamento por Data de Referência')
plt.xlabel('Data de Referência')
plt.ylabel('Faturamento (em R$)')

ax.set_ylim(0, df.valor.max()) 
```

```python
ax = df.plot.line(x='data_ref', y='valor', rot=0, figsize=(15,6), marker='o')

plt.title('Faturamento por Data de Referência')
plt.xlabel('Data de Referência')
plt.ylabel('Faturamento (em R$)')

plt.yticks([0, 200000, 400000, 600000, 800000, 1000000], 
           ['0', '200k', '400k', '600k', '800k', '1M'])
plt.show()
```
